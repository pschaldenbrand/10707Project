# Convert the MP3 files to WAV
Wave files are easier to deal with than mp3
- Convert the mp3 files into wav in a similar directory structure
- Ensure that the audio files are mono aka single channel for ease later

```python
import os
rootdir = 'fma_small'

outputDir = 'wav'

if not os.path.exists(outputDir):
    os.makedirs(outputDir)
```

```python
from pydub import AudioSegment
```

```python
MAX_SONGS = 9000

numSongs = 0

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        try:

            if not file.endswith(".mp3"):
                continue
            if (numSongs %100 == 0):
                print(numSongs)
            numSongs += 1
            if (numSongs > MAX_SONGS):
                break

            track_id = int(os.path.splitext(file)[0])

            sound = AudioSegment.from_mp3(os.path.join(subdir, file))
            if sound.channels == 2:
                [left, right] = sound.split_to_mono()
                sound = right
                sound.set_channels(1)

            newDir = "wav/" + subdir.split("\\")[1] # Change delimiter to "/" if mac
            if not os.path.exists(newDir):
                os.makedirs(newDir)
            sound.export(newDir +"/"+ os.path.splitext(file)[0] + ".wav", format="wav")

        except Exception as e: 
            print('Error with this file', e)
    if (numSongs > MAX_SONGS):
        break
```
